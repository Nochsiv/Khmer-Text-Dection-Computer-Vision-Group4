```python
from ultralytics import YOLO

# Load the YOLOv11 model
model = YOLO("yolo11n.pt")
```

```python

# Train the model with the --single-cls flag
yaml = "/teamspace/studios/this_studio/yolo11/Scence-Text-Detection-4/data.yaml"
results = model.train(data=yaml, epochs=100, imgsz=640, single_cls=True)
```

```python
# Print the training results
print(results)
```

```python
model.save("yolo11n_100p.pt")
```
