```python
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("yolov5x6u.pt")  # Load a specific YOLOv8 model (e.g., yolov8s, yolov8m, yolov8l)

# Train the model with the --single-cls flag
results = model.train(data="data.yaml", epochs=100, imgsz=640, single_cls=True)

# Print the training results
print(results)
```

```python
model.save("trained_yolov5x6u.pt")
```
